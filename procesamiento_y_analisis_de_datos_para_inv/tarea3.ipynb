```python
import numpy as np
import pandas as pd
```

```python
datos = pd.read_excel("tarea 3.xlsx",sheet_name="tabla dinámica")
```

```python
datos.columns
```

```python
datos.describe()
```

#### Ventas por persona según edad

```python
pd.pivot_table(datos,values="ventas",index="nombre",columns="edad",aggfunc=np.sum,fill_value=0.0)
```

##### Por edad

```python
datos.groupby("edad")[["ventas"]].sum()
```

#### Ventras por estrato y sexo segun edad

```python
pd.pivot_table(datos,values="ventas",index=["estrato","sexo"],columns="edad",aggfunc=np.sum)
```

#### Ventas solo para hombres por ciudad

```python
datos_hombres = datos[datos.sexo == "masculino"]
datos_hombres.drop(["edad","estrato"],axis = 1).groupby("ciudad").agg("sum")
```
