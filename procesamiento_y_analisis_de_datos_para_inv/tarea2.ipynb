### Estadistica descriptiva basica
Medias, desviacion estandard, percentiles(25,50,75), min y max por olumna

```python
datos.describe()
```

### Promedio Agrupado por Seccion

```python

datos.drop(["No.","Sexo"],axis=1).groupby("Secc").agg(["mean"])
```

### Minimo por Seccion

```python
datos.drop(["No.","Sexo"],axis=1).groupby("Secc").agg(["min"])
```

### Maximo por seccion

```python
datos.drop(["No.","Sexo"],axis=1).groupby("Secc").agg("max")
```

### Cantidad de sujetos

```python
datos.drop(["No.","Sexo"],axis=1).groupby("Secc").agg("count")
```
